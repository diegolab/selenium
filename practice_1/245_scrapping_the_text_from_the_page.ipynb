```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome()
driver.maximize_window()
url = "https://omayo.blogspot.com"
driver.get(url)
```

```python
text_on_page = str(driver.execute_script("return document.documentElement.innerText"))
```

```python
print(text_on_page)
```
